Principios de Comunicações - Modulação AM e sobremodulação


A equação para modulação AM é definida por: 

$s(t) = A_c [1 + k_a m(t)] cos(2 \pi f_c t)$

Nessa equação temos que $s(t)$ representa o sinal após a modulação. $A_c$ e $f_c$ representam respectivamente a amplitude e a frequência da portado. 

Além disso o mais importante são os termos $k_a$ e $m(t)$, no qual $m(t)$ é o sinal original que contém a informação que desejamos modular, e  $k_a$ é uma constante chamada de sensibilidade de amplitude do modulador responsável pela geração do sinal modulado.

Só temos modulação presente quando $|k_a m(t)| < 1$

Caso essa premissao não seja verdadeira, ou seja, temos que $k_a$ é maior do que $1$, nos obtemos um sinal sobremodulado, que é o objetivo dessa atividade.  

No primeiro caso vamos observar um sinal modulado, onde $k_a$ é igual a $0$.


```python
import numpy as np
import matplotlib.pyplot as plt

amostras = 1000
amp = 1
fc = 5
k = 0

s = []
for t in range(amostras):
    m_t = np.cos(2 * np.pi * t / 1000)  # Correção na definição de m_t
    s.append(amp * (1 + m_t * k) * np.cos(2 * np.pi * (t / 1000) * fc))

plt.plot(s)
plt.ylim(-3, 3)
plt.show()
```
