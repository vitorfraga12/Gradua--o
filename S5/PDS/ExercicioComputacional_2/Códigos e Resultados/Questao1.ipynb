```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import sounddevice as sd
```

```python
#    
bipsIN, fs = librosa.load('bipsIN.wav', sr=None)
period = 1/fs
time = np.arange(0, len(bipsIN)*period, period)

plt.plot(time, bipsIN)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.savefig('q1_bipsIN_time.pdf')
plt.show()

print(period)
print(fs)
```

A fórmula padrão para amostragem de um sinal é dada por 

$x[n] = x_c(nT)$    

Podemos também representar por um trem de impulsos, ficando então:

$x_s(t) = x[n] = \sum_{n = -\infty}^{\infty} x_c(nT)\delta(t-nT) $

Consideramos que $T = \frac{1}{F_s} \therefore T = \frac{1}{5000} = 0.0002  $

Podemos então inferir que a equação que iremos obter é:

$x[n] = \sum_{n = 0}^{25000} x_c(0.0002n)\delta(t - 0.0002n)      $




```python
# Using the fourrier transform, plot the magnitude of the frequency spectrum of the sound file bipsIN.wav

fft_bipsIN = np.fft.fft(bipsIN)
freq = np.fft.fftfreq(len(fft_bipsIN), 1/fs)
freq = freq[:len(freq)//2]	
fft_bipsIN = fft_bipsIN[:len(fft_bipsIN)//2]
plt.plot(freq, np.abs(fft_bipsIN))
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.savefig('q1_bipsIN_freq.pdf')
plt.show()
```

Já para a equação na frequência, temos que:

$ X(e^{j\omega}) = \sum_{n = -\infty}^{\infty} x[n]e^{-j\omega n}  $

Podemos reescrever essa equação da seguinte maneira:

$ X(e^{j\omega}) = \frac{1}{T}\sum_{k=-\infty}^{\infty} X_c [j(\frac{\omega}{T} - \frac{2\pi k}{T})]  $

Dessa maneira, a representação desse sinal é:

$ X(e^{jw}) = \frac{1}{0.0002}\sum_{k=0}^{2500} X_c[j(\frac{\omega}{0.0002} - \frac{2\pi k}{0.0002})] $ $\therefore $ $ X(e^{jw}) = 5000\sum_{k=0}^{2500} X_c[j(\frac{\omega}{0.0002} - \frac{2\pi k}{0.0002})]  $
